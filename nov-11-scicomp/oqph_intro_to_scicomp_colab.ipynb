#### 3.2.2 Inner Product

The inner product of a vector results into a scalar value. The purpose of this operation is to know the similarity of two different vectors through directional multiplication. The inner product of two vectors can be represented by the equation:
$$\sum^N_{i=0}u_i\times v_i$$
Whereas $u$ and $v$ are vectors wherein they both have $N$ number of elements. This can be further simplified using the inner product vector form:
$$u\cdot v$$
Again still following the rule that $u$ has the same size as $v$.

```python
C = np.array([1,2,3])
X = np.array([0,-1,0.5])
```

```python
V = np.inner(C,X)
V
```

```python
V = C @ X
V
```

## Part 5: Note-worthy Techniques

```python
def create_linear_trend(n,is_neg=False):
  inv = -1 if is_neg else 1
  X = np.arange(0,n,1,dtype=np.float64) ## features
  m = inv*np.random.uniform(0.1,0.3,(n,)) ## weights
  b = np.random.uniform(8,10,(n,)) ## biases
  y = m*X+b  ## targets
  return X,y,m,b
```

```python
n=100
X,y,_,_ = create_linear_trend(n,is_neg=True)

plt.figure(figsize=(5,5))
plt.grid()
plt.scatter(X,y)
plt.show()
```

```python
def create_gaussian_2d(n, center, covar):
  np.random.seed(0)
  cov_mat = np.array([
                    [1,covar],
                    [covar,1]
                    ])
  
  gauss2d = np.array(np.random.multivariate_normal(center ,cov_mat,int(n/2)))
  gaussX = gauss2d[:,0]
  gaussY = gauss2d[:,1]
  return gaussX, gaussY
```

```python
c = np.array([2, 0]).T
gX, gY = create_gaussian_2d(n=500, center=c, covar=0.7)
plt.scatter(gX, gY)
```

### 5.1 Pearson's Correlation

Another measure of relationship of datasets is the Pearson Correlation Coefficient or Pearson's R. Pearson's R is specifically a measure of linear correlation between data. It can be interpreted as the ratio between the covariances of the two variables and the product of their standard deviations.
$$r_{X,y} = \frac{\text{Cov(X,y)}}{\sigma_X \sigma_y} \\_{\text{Eq. }}$$

```python
n=100
line_x,line_y,_,_ = create_linear_trend(n,is_neg=False)

plt.figure(figsize=(5,5))
plt.grid()
plt.scatter(line_x,line_y)
plt.show()
```

```python
np.corrcoef(line_x, line_y)
```

```python
c = np.array([2, 0]).T
gauss_x, gauss_y = create_gaussian_2d(n=500, center=c, covar=0.5)
plt.scatter(gauss_x, gauss_y)
```

```python
np.corrcoef(gauss_x, gauss_y)
```

### 5.2 Least Squares

We will first use the **Normal Equation** in solving linear regression. The Normal equation uses the Least-Squares Cost function and is formulated as:
$$\theta = (X^TX)^{-1}X^Ty \\ _{\text{(Eq. 4.5.2)}}$$
Whereas $\theta$ is the hypothesis or model to be created while $X$ represents the data vector and $y$ represents the labels or values corresponding to the data vector. The term $(X^TX)^{-1}$ is called the **pseudoinverse** or the **Moore-Penrose** matrix. The pseudoinverse of a matrix term of Eq. 4.5 assures that the data are normal or orthogonal. This helps check the property of Autocorrelation between the features of the data. The other properties of datasets that are safe for linear regression are Homoscedasticity, Non-multicollinearity, and Non-endogeneity. 

```python
X = np.asarray(life_expect_predictors.drop(labels=['life_expectancy'], axis=1))
bias = np.ones((X.shape[0],1))
X = np.append(X, bias, axis=1) ## add a bias term
y = np.asarray(life_expect_predictors['life_expectancy'])
```

```python
theta = np.linalg.inv(X.T@X)@X.T@y
theta
```

```python
from sklearn.linear_model import LinearRegression
X = np.asarray(life_expect_predictors.drop(labels=['life_expectancy'], axis=1))
linear_reg = LinearRegression()
linear_reg.fit(X,y)
```
